```python
# removal of punctuation : 

PUNCT_TO_REMOVE = string.punctuation                                             # these are punctuations,   !"#$%&\'()*+,-./:;<=>?@[\\]^_{|}~`   we have to choose which to remove and retain
def remove_punctuation(text):
    """custom function to remove the punctuation, we """
    return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))
```

```python
# removal of stopwords: 
''' In cases like Part of Speech tagging, we should not remove them as provide very valuable information about the POS. And also in nltk.corpus words like "not","no" comes under stopwords we should remove them from corpus in sentiment analysis taks'''

", ".join(stopwords.words("english"))


```

```python
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    """ custom function to remove the stopwords """
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])
```

```python
sentence = " Machine Learning is one of the most prominent area of Technology in 21st century"

remove_stopwords(sentence)
```

```python
# Removal of frequent words  ( this is in built when we use tf-idf for vectorization)

count = Counter()
# for data in df["<text_column_name>"].values:                                                               # use this when dealing with columns of text data in dataframe
for word in data.split():
    count[word]+=1
count.most_common(10)
```

```python
FREQWORDS = set([w for (w, wc) in count.most_common(10)])
def remove_freqwords(text):
    """custom function to remove the frequent words"""
    return " ".join([word for word in str(text).split() if word not in FREQWORDS])
```

```python
remove_freqwords(data)
```

```python
count.most_common()[-10:] 
```

```python
# removal of Rare Words: similar to prev method with small changes

RAREWORDS = set([w for (w, wc) in count.most_common()[-10:]])
def remove_rarewords(text):
    """custom function to remove the frequent words"""
    return " ".join([word for word in str(text).split() if word not in RAREWORDS])
```

```python
remove_rarewords(data)
```

```python
# Stemming  : the process of reducing the words to their base words, ex: joined, joining -> join

stemmer = PorterStemmer()

def stem_words(text):
  new_text = " ".join([stemmer.stem(word) for word in text.split()])
  return new_text
```

```python
stem_words(data)
```

```python
stemmer.stem("private")
```

```python
# lemmatization: similar to stemming and more meaninigful but slow camparitively

lemmatizer = WordNetLemmatizer()

def lemmatize_words(text):
   new_text = " ".join([lemmatizer.lemmatize(word) for word in text.split()])
   return new_text
```

```python
lemmatize_words(data)
```

```python
lemmatizer.lemmatize("private")
```

```python
lemmatizer = WordNetLemmatizer()
wordnet_map = {"N":wordnet.NOUN, "V":wordnet.VERB, "J":wordnet.ADJ, "R":wordnet.ADV}
def lemmatize_tagged_words(text):
    pos_tagged_text = nltk.pos_tag(text.split())
    return " ".join([lemmatizer.lemmatize(word, wordnet_map.get(pos[0], wordnet.NOUN)) for word, pos in pos_tagged_text])
```

```python
lemmatize_tagged_words(data)
```

```python
# Removing emojis, is not a good idea to do when doing sentiment tasks. Again this step is task specific

# Reference : https://gist.github.com/slowkow/7a7f61f495e3dbb7e3d767f97bd7304b
def remove_emoji(string):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emojis 
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)
```

```python
def remove_emoticons(text):
    emoticon_pattern = re.compile(u'(' + u'|'.join(k for k in EMOTICONS) + u')')
    return emoticon_pattern.sub(r'', text)
```

```python
remove_emoticons("Hello :-)")
```

```python
# converting Emoticons to words 

def convert_emoticons(text):
    for emot in EMOTICONS:
        text = re.sub(u'('+emot+')', "_".join(EMOTICONS[emot].replace(",","").split()), text)
    return text
```

```python
text = "Hello :-) :-)"
convert_emoticons(text)
```

```python
# Removal of urls 

def remove_url(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)  
```

```python
chat_words_map_dict = {}
chat_words_list = []
for line in chat_words_str.split("\n"):
    if line != "":
        cw = line.split("=")[0]
        cw_expanded = line.split("=")[1]
        chat_words_list.append(cw)
        chat_words_map_dict[cw] = cw_expanded
chat_words_list = set(chat_words_list)

def chat_words_conversion(text):
    new_text = []
    for w in text.split():
        if w.upper() in chat_words_list:
            new_text.append(chat_words_map_dict[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)
```

```python
# Spell chekcing 

from spellchecker import SpellChecker

spell = SpellChecker()
def correct_spellings(text):
    corrected_text = []
    misspelled_words = spell.unknown(text.split())
    for word in text.split():
        if word in misspelled_words:
            corrected_text.append(spell.correction(word))
        else:
            corrected_text.append(word)
    return " ".join(corrected_text)
```

# Above was only for text data feature, for Categorical related features steps are different


### in categorical features sometimes the names of categories are different like , they will be having space or - or some other things in the category names. It is important to preprocess it 

### for example consider,
Grades PreK-2    
Grades 3-5       
Grades 6-8       
Grades 9-12      , 
these are the cateogies in a feature. But they have space and "-" in between thus while encoding them it creates a problem 

we have preprocess them by removing " ", "-"  and replace them with "_" 

Thus forming

grades_prek_2 

grades_3_5 

grades_6_8 

grades_9_12 

2. For features like latitude and longitude we can take a common point and encode latitude and longitude values as distance from that common point( say equator or Bangalore latitude and longitude)

3. For features like zip code, we can one-hot 

4. for features like Id of indivisual,  there is not much we can do about it . Since there is not relation between any 2 values and also they are not ordinal. They cant be referred from common point like in latitude and longitude case.

Most of the time we remove these Id Features

### After getting the final processed text, we do furthur feature engineering if there are any depending on task, and feature selection if needed. 
### Next we encode (tf-df, word2vec or embedding) the features to Vector form and train the models
