```python
dataset_root = '/content/lfw_faces/lfw_funneled'
count = 0

def preprocess_face(img_arr):
  ''' standardized the image and expands axis'''
  img_arr = img_arr.astype('float')
  mean,std = img_arr.mean(),img_arr.std()
  img_arr = (img_arr - mean)/std
  img_arr = np.expand_dims(img_arr, axis=0)
  return img_arr

def get_features_labels(dir):
  ''' accepts directory name and returns face embeddings with labels'''
  image_path = os.path.join(dataset_root, dir)
  face_embed = []  
  global count
  count+=1

  for img in os.listdir(image_path):
    image = Image.open(os.path.join(image_path, img))
    image = image.convert('RGB')
    img_arr = np.asarray(image)
    res = detect_model.detect_faces(img_arr)                                    # returning face cordinates
    if len(res)>0:
      x, y, width, height = res[0].get('box')
      face = img_arr[y:y+height, x:x+width]
      face_img = Image.fromarray(face)
      face_img = face_img.resize((160,160))                                     # resizing face image to (160,160) required by Facenet
      face_arr = np.asarray(face_img)
      face_arr = preprocess_face(face_arr)
      face_embed.append(facenet.predict(face_arr)[0])                              # extracting face embeddings
  
  label = [count]*len(face_embed)
  return face_embed, label     
```

```python
# testing for euclidian distance between different indivisuals and between same indivisual ( to set proper threshold )

def get_euclidian_distance(embed1, embed2):
  normalizer = Normalizer(norm='l2')
  embed1 = normalizer.transform(embed1.reshape(1,-1))
  embed2 = normalizer.transform(embed2.reshape(1,-1))
  diff = embed1 - embed2
  res = np.linalg.norm(diff)
  return res


# for same person : 
x,y = get_features_labels('Eric_Rosser')
print(" Euclidian distance between same person is : ", get_euclidian_distance(x[0], x[1]))

# for same person :
x,y = get_features_labels('Stacy_Dragila')
print(" Euclidian distance between same person is : ", get_euclidian_distance(x[0], x[1]))

# euclidian distance between different persons 
x1, y1 = get_features_labels('Kitin_Munoz')
x2, y2 = get_features_labels('Neil_Goldman')
print(" Euclidian distance between same person is : ", get_euclidian_distance(x1[0], x2[0]))

# euclidian distance between different persons 
x1, y1 = get_features_labels('Tom_OBrien')
x2, y2 = get_features_labels('Lee_Soo-hyuck')
print(" Euclidian distance between same person is : ", get_euclidian_distance(x1[0], x2[0]))

```

```python
# to improve the above accuracy, getting the neighbours and classifying based on threshold conditions
neigh = NearestNeighbors(n_neighbors=3)
neigh.fit(x_train)
```

```python
def get_neighbours(model, query):
  res = model.kneighbors(query, return_distance=True)
  return res
```

```python
get_neighbours(neigh, x_test[77].reshape(1,-1))
```

```python
get_neighbours(neigh, x_test[2].reshape(1,-1))
```

```python
threshold = 0.6
predicted = []

# for ones whose distance<=threshold predict directly, ones > threshold take distance between 2 min points and check if their distance is less than max distance for query point

for person in tqdm(x_test):
  distances, indices= get_neighbours(neigh, person.reshape(1,-1))
  if min(distances[0])<=threshold:
    index_of_min = np.where(distances[0]==min(distances[0]))
    pred_index = indices[0][index_of_min]
    predicted.append(y_train[pred_index].tolist()[0])
    continue
  else:
    sorted_indices = np.argsort(distances[0])    
    # mean_distance = sum(distances[0][sorted_indices[:2]])/2
    # print(mean_distance)
    two_neighbours = sorted_indices[:2]
    near_points = indices[0][two_neighbours]
    first_point, second_point = x_train[near_points[0]], x_train[near_points[1]]
    distance_between = np.linalg.norm(first_point-second_point)
    if distance_between <= threshold or distance_between<= max(distances[0]):
      predicted.append(y_train[near_points[0]])
    else:
      predicted.append(-1)      
 
```

```python
count = 0
for i in range(len(y_test)):
  if y_test[i]==predicted[i]:
    count+=1

print("Accuracy : ", count/len(y_test))
```

```python
print("Precision Score is : ",precision_score(y_test, predicted, average='micro'))
```
