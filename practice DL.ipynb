```
from keras.datasets import reuters
import tensorflow as tf
from keras.utils import to_categorical
from keras import models
from keras import layers
from keras import regularizers
```

```
(x_train,train_label),(x_test,test_label) = reuters.load_data(num_words=10000)
```

```
np.min(train_label)
```

```
test_label = to_categorical(test_label)
```

```
test_label[0]
```

```
test_label.shape
```

```
train_label = to_categorical(train_label)
```

```
train_label.shape
```

```
# padding and vectorizing the data
import numpy as np

def vectorize(data):
  res = np.zeros((len(data),10000))
  for index,news in enumerate(data):
    res[index,news]=1
  return res  
```

```
x_train = vectorize(x_train)
x_test = vectorize(x_test)
```

```
x_test.shape
```

```
train_label = to_categorical(train_label)
test_label = to_categorical(test_label)
train_label.shape
```

```
train_label[0]
```

```
model = models.Sequential()
```

```
model.add(layers.Dense(70,input_shape=(10000,),activation="relu"))
model.add(layers.Dense(50,activation="relu"))
model.add(layers.Dense(46,activation="softmax"))
```

```
model.compile(loss="categorical_crossentropy",metrics="accuracy",optimizer="adam")
```

```
model.fit(x_train,train_label,validation_data=(x_test,test_label),epochs=20,batch_size=100)
```

clear overfitting

```
# adding regularizer to avoid overfitting 
```

```
test_label.shape
```

```
model = models.Sequential()
model.add(layers.Dense(70,input_shape=(10000,),activation="relu",kernel_regularizer=regularizers.l2(0.01)))
model.add(layers.Dense(50,activation="relu",kernel_regularizer=regularizers.l2(0.001)))
model.add(layers.Dense(46,activation="softmax"))
model.compile(loss="categorical_crossentropy",metrics="accuracy",optimizer="adam")
```

```
model.fit(x_train,train_label,validation_data=(x_test,test_label),epochs=20,batch_size=100)
```

```
# it still overfitting
```

```
# reducing the model capacity to make sure it memorizes only the most important things in train data
```

```
model = models.Sequential()
model.add(layers.Dense(80,input_shape=(10000,),activation="relu",kernel_regularizer=regularizers.l2(0.01)))
model.add(layers.Dense(55,activation="relu",kernel_regularizer=regularizers.l2(0.01)))
model.add(layers.Dense(46,activation="softmax"))
model.compile(loss="categorical_crossentropy",metrics="accuracy",optimizer="adam") 
```

```
model.fit(x_train,train_label,validation_data=(x_test,test_label),epochs=20,batch_size=50)
```

```
 # adding dropouts to regularize more 
```

```
model = models.Sequential()
model.add(layers.Dense(80,input_shape=(10000,),activation="relu",kernel_regularizer=regularizers.l2(0.01)))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(55,activation="relu",kernel_regularizer=regularizers.l2(0.01)))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(46,activation="softmax"))
model.compile(loss="categorical_crossentropy",metrics="accuracy",optimizer="adam") 
```

```
model.fit(x_train,train_label,validation_data=(x_test,test_label),epochs=20,batch_size=50)
```
