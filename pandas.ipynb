```python
import pandas as pd
```

```python
x = pd.date_range('1997-12-30', '2021-09-12').to_series()
```

```python
res = {
    "day":x.dt.day,
    # "day_name":x.dt.day_name,
    "year":x.dt.year,
    "month":x.dt.month
}
```

```python
df = pd.DataFrame(res)
```

```python
df.head()
```

```python
df.groupby("year").describe()
```

```python
df.groupby("year").apply(print)
```

```python
df.groupby("year").count()
```

```python
df.groupby("year").last()
```

```python
df["year"].unique()
```

```python
df.groupby("year").count()    # basically the aggregate or any operation will be on numeric columns and except the column on which we are grouping since it acts as index column
```

```python
df.groupby("year").agg(["min","max"])
```

```python
df.groupby("year").agg({"day":"count", "month":["min","max"]})  # cases where we want to apply different aggregation operation on different columns
```

```python
df.groupby("year").agg({"day":"count"})    
```

```python
df.groupby('year').agg({"year":"count"})           # we can use 
```

```python
for key,val in df.groupby("year"):
  print(key,val)
```

```python
df.groupby("year").agg({"cat1":"count","cat2":"count"}).
```

```python
from functools import partial 
```

```python
def add(a,b,c):
  return a+b+c
```

```python
new_fun = partial(add, a=1,b=2)
```

```python
print(new_fun(c=3))
```

```python
new_fun = partial(add, a=10, c=20)
```

```python
print(new_fun(b=30))
```
